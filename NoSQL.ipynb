# NoSQL

`Not Only SQL`

It's a type of database designed for:
- Flexible data.
- Fast scaling.
- Huge amounts of information.
- Modern websites/apps.

Unlike SQL databases, NoSQL databases do `not always use tables and rows.`    

## SQL vs NoSQL

`SQL` `-`  `NoSQL`
- Tables                               - Flexible structures

- Rows and columns                     - Documents or key-values or graphs

- Fixed schema                         - Dynamic schema

- Structured data                      - Unstructured data

- SQLite, MySQL                        - MongoDB, Firebase 

- Best for banking                     - Best for apps or social media

- Scalabilty(vertical-big server)      - (Horizontal- more servers)

### REAL-WORLD EXAMPLES

`SQL`

- Best for: `hospitals`, `banks`, `school systems`, `payroll`.

- This is because `relationships matter`.


`NoSQL`

- Best for: `chat apps`, `social media`, `analytics`, `AI apps`, `recommendation systems`.

- This is because `data changes constantly`.

### MAIN TYPES OF NoSQL

`1. Document Databases`

- Stores data like JSON documents.
- This is powerful because different users can have different fields.
- Also no need for identical columns.
- Example: `MongoDB`

Instead of tables, example:


```python
# JSON
{
    "name": "Tatiana",
    "age": 21,
    "mood_score": 7,
    "interests": ["Lifestyle", "Mental Health"]
}
```

`2. Key-Value Databases`

- It's very fast.
- Used for: `caching`, `sessions`, `settings`.
- Example: `Redis`

example:


```python
user123 - "Tony"
theme - "light"
```

`3. Graph Databases`

- Stores relations.
- Used for: `social networks`, `recommendations`.
- Example: `Neo4j`

```python
Peter - friends with - Sarah 
Sarah - follows - Michael
```

`4. Column Databases`

- Stores data by columns instead of rows.
- Used for: `analytics`, `huge datasets`.
- Example: `Cassandra`
