#### Replacing parts of a String 

`.replace(text_to_be_replaced , text_to_change_it_to)`

```python
colour_statement = "Isaac's favourite colour is blue. He also likes green but prefers blue."
print(colour_statement)
```

```python
right_colour_statement = colour_statement.replace("brown", "blue", -1)
print(right_colour_statement)
```

# Data Structures In Python

## Lists in Python

- Lists store multiple values in a single variable
   - Lists can contain any combination of data type eg `boolean` , `string` , `int`

- Lists are ordered
   - One can use subsetting or indexing
   -python starts counting from `zero` for the first element   

```python
fuel_prices = [90, 150, 100, 194, 60, 76]
print(fuel_prices)
```

### Accessing Elements of a List

- We use the square brackets to subset - []
- Sub-setting syntax : `name_of_list[index]`

```python
fuel_prices[2]
```

```python
fuel_prices[-4]
```

#### Slicing in Lists / Accessing Multiple Elements in a List

```python
fuel_prices
```

```python
fuel_prices[2:6]
```

```python
# Get the first three fuel prices
# These are values from index zero to index two (not inclusive of index three)
fuel_prices[:3]
```

```python
# Access all the elements from the fourth index onwards
fuel_prices[3:]
```

##### Alternating Access

```python
fuel_prices
```

```python
# Accessing Every other element in the list
fuel_prices[::2]
```

```python
# Accessing every third element in the list starting from the second element(index 1)
fuel_prices[1::3]
```

### Definition

- A `Tuple` is a data type that stores a collection of items.
- `Tuples` are used to store multiple items in a single variable.
- It is `ordered` and `immutable`.

### Immutability Of Tuples

- They do not allow modification.
- You cannot change, add or remove any value after creation.

```python
# Example : Trying to change a value in a tuple

numbers = (1, 2, 3, 4, 5)
numbers[0] = 10  # This will raise a TypeError because tuples are immutable.
```

```python
number = [1, 2, 3, 4, 5]
number[0] = 10  # This will work because lists are mutable.
print(number)
```

```python
# Example : Trying To Add or Remove items in a Tuple

numbers = (1, 2, 3, 4, 5)
numbers.append(6)  # This will raise an AttributeError because tuples do not have an append method.
```

```python
# Example : Trying To Delete a Single Item in a Tuple

numbers = (1, 2, 3, 4, 5)
del numbers[0]  # This will raise a TypeError because tuples do not support item deletion.
```

### Methods That Apply In Tuples

#### 1. count() method

- This method counts the number of times a specified value appears in the tuple.



```python

# Example
numbers = (1, 2, 3, 4, 5, 2, 2)
print(numbers.count(2))  # 3
```

#### 2. index() method

- This method returns the index of the first occurrence of a specified value in a tuple. 

```python
# Example
numbers = (1, 2, 3, 4, 5, 2, 2)
print(numbers.index(3))  # 2
```

## Sets In Python

- Contains Unique Data
- They are Unchangable
     - You can add or remove but you can't change what's already there

`set = {}`

```python
mokimo_ingredients = {"Potatoes", "Pumpkin Leaves", "Margarine", "Salt", "Peas",
                       "Onions", "Tomatoes", "Carrots", "Peppers"}
print(mokimo_ingredients)
```

```python
mokimo_ingredients_list = ["Potatoes", "Pumpkin Leaves", "Margarine", "Salt", "Peas",
                       "Onions", "Tomatoes", "Carrots", "Peppers"]
print(mokimo_ingredients_list)
```

```python
# Changing a List to  be a Set
mokimo_ingredients_from_list = set(mokimo_ingredients_list)
print(mokimo_ingredients_from_list)
```

```python
# Checking the type of the variable
print(type(mokimo_ingredients_from_list))
```

```python
# Sorting my Set
sorted_mokimo_ingredients = sorted(mokimo_ingredients_from_list)
print(sorted_mokimo_ingredients)
```

```python
print(type(sorted_mokimo_ingredients))
```
