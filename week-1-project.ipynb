# CLI Quiz Game Application

```python
# This is a simple command-line based quiz application.
# It allows user to choose category, answer multiple-choice questions,
# checks answers then displays final score,
# and gives feedback based on performance.

import random

# List of questions.
questions = {
    "biology": [
        {
            "question": "Which organelle is responsible for cellular respiration?",
            "choices": ["A. Ribosome", "B. Chloroplast", "C. Mitochondrion", "D. Golgi Apparatus"],
            "answer": "C"
        }
    ],
    "chemistry": [
        {
            "question": "What is the chemical formula for Sodium Chloride?",
            "choices": ["A. NaCl", "B. KCl", "C. HCl", "D. Mg(OH)Cl"],
            "answer": "A"
        }
    ]
}

# Choose category
print("Choose a category:")
print("i. Biology")
print("ii. Chemistry")

option = input("Enter i or ii: ").strip().lower()
if option == "i":
    selected_questions = questions["biology"]
elif option == "ii":
    selected_questions = questions["chemistry"]
else:
    print("Invalid choice, defaulting to Biology.")
    selected_questions = questions["biology"]

random.shuffle(selected_questions)

score = 0

# Loop through questions
for q in selected_questions:
    print("\n" + q["question"])
    for choice in q["choices"]:
        print(choice)

    answer = input("Enter answer: ").strip().upper()

    if answer == q["answer"]:
        print("Correct")
        score += 2
    else:
        print("Wrong")

# Final result
print(f"\nYour Score: {score}")

# Feedback
max_score = len(selected_questions) * 2
if score == max_score:
    print("EXCELLENT")
elif score >= 1:
    print("GOOD JOB")
else:
    print("TRY AGAIN")
        

```
