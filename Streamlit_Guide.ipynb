# STREAMLIT


- `Streamlit` is a Python framework that turns Python scripts into `interactive web apps/ dashboard websites`.

- Used for: Dashboards, Data science apps, AI tools, Analytics platforms, Surveys, Porfolio projects.


##### INSTALL STREAMLIT

- `pip install streamlit`

##### CREATE YOUR FILE

- Example: `app.py`

##### BASIC STREAMLIT APP

import streamlit as st

st.title("FoodPlus")

st.write("Welcome to your AI-powered recipe app")

##### RUN IT

- `streamlit run app.py` - streamlit opens [https://localhost:8501] and that becomes your website.


- Streamlit command is `st.`

## STREAMLIT BUILDING BLOCKS

### ` 1.TITLES`

st.title("Main Title")

st.header("Header")

st.subheader("Subheader")

### `2.TEXT`

st.write("Normal text")

st.text("Plain text")

st.markdown(**Bold text**)

### `3.BUTTONS`

if st.button("Click Me")

 st.write("Button Clicked!")

### `4.USER INPUT`

name = st.text_input("Enter your name:")

st.write("Hello", name)

### `5.DROPDOWN`

food = st.selectbox("What food would you like to order?, ")

st.write("You ordered: ", food)

#### `6.SLIDER`

stress = st.slider("Stress Level", 1, 10)

st.write("Stress Score:", stress)

#### `CODE EXAMPLE:`

import streamlit as st

st.title("MindPulse Check-In")

name = st.text_input("Your Name")

sleep = st.slider("Hours of Sleep", 0, 12)

stress = st.slider("Stress Level", 1, 10)

if st.button("Analyze"):

st.write(f"{name}, you slept {sleep} hours, your stress level is {stress})
