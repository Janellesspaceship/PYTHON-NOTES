Conditional statements are statements that perform a specific task if the condition is `True` or if `False` do nothing.

`if` > condition is met > perfom action

```python
# As a vendor at a Concert you have a tatrget number of units you should sell

sales_target = 300 
units_sold = 279

# Let me compare the sales

if units_sold >= sales_target:
    print("Target Achieved")

 # Check if we are close to the target
elif units_sold >= 250: 
    print("Target almost achieved")

 # Otherwise....
elif units_sold == 290:
    print("10 units left to hit target")

elif units_sold == 279:
    print("279 units sold, 21 units left to hit target")

else:
    print("Target not achieved")  
```

- You can use as many `elif` statements as you like

# For Loops

```python
# Customer Feedback Scores
feedback_scores = [30, 45.0, 55.9, 60, 30.56, 91.1, 75.6, 80.6, 12.9]
```

```python
# Check if each score is above 50
feedback_scores[6] > 50
```

- `for` each value in a `sequence` , perform `action`
    - `action` is indented because of the colon in the previous line

- `sequence` = iterable eg: List, Dictionary etc
    
- `value` = iterator which is the index
     - Placeholder (You can give `value` any name) but `i` is common.

```python
# Print each value in the List Feedback Scores

for score in feedback_scores:
    print(score)
```
