Conditional statements are statements that perform a specific task if the condition is `True` or if `False` do nothing.

`if` > condition is met > perfom action

```python
# As a vendor at a Concert you have a tatrget number of units you should sell

sales_target = 300 
units_sold = 279

# Let me compare the sales

if units_sold >= sales_target:
    print("Target Achieved")

 # Check if we are close to the target
elif units_sold >= 250: 
    print("Target almost achieved")

 # Otherwise....
elif units_sold == 290:
    print("10 units left to hit target")

elif units_sold == 279:
    print("279 units sold, 21 units left to hit target")

else:
    print("Target not achieved")  
```

- You can use as many `elif` statements as you like

### Sample Keywords in Python

- `in` = Checks if a value is in a variable or a Data Structue(List, Tuple, Dictionary etc)

- `not` = Checks if a condition is not met

- `and` = Checks if multiple conditions are met

- `or` = Checks if one ( **Or more** ) condition is met

```python
# The `in` Keyword

ArtemisII_Astronauts = {"Reid" : {"Age": 39, "Role": "Commander"},
                        "Charlie" : {"Age": 42, "Role": "Pilot"},
                        "Kim" : {"Age": 25, "Role": "Engineer"}}

```

```python
# Print the keys of the `ArtemissII_Astronauts` dictionary
print(ArtemisII_Astronauts.keys())
```

```python
# Print the values ot the `ArtemissII_Astronauts` dictionary
print(ArtemisII_Astronauts.values())
```

```python
# Use the `and` keyword to check if "Kim" is a key in the `ArtemissII_Astronauts` dictionary 
# and if the minimum age of all the astrounats is> 20

if "Kim" in ArtemisII_Astronauts.keys() and min([astronaut['Age'] for astronaut in ArtemisII_Astronauts.values()]) > 20:
    print(True)
    
else:
    print(False)
```

```python
# Use the `or` keyword to check if Charlie is a key in the `ArtemissII_Astronauts` dictionary 
# and if the minimum age of all the astrounats is> 20

if "Charlie" in ArtemisII_Astronauts.keys() or min( ArtemisII_Astronauts.values()) > 20:
    print(True)
else:
    print(False)    
```
