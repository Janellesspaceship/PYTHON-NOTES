## Regular Expressions

- A `RegEx`, or `Regular Expression`, is a sequence of characters that forms a search pattern.

- RegEx can be used to check if a string contains the specified search pattern.

##### Python Module For RegEx
- Import the `re` module:

   `re`

- A way to describe a pattern in text.

### How We Use RegEx In Python

Functions:

`re.search()` - finds pattern anywhere returns the first match found

`re.match()`  - checks for pattern only at the start of a string

`re.sub()` - finds pattern and replaces it with something else (substitute)

`re.findall()` - returns a list containing all matches of a pattern in a string



### Why We Use RegEx

- Because normal string methods are limited.

##### 1. Search Complex Patterns

```python
# Working With Structured Text e.g. emails, numbers, logs
# re.search() - Search for a pattern in a string and return the first match

import re

text = "My number is 0709832707"
result = re.search(r"\d{10}", text)
if result:
    print("Phone number found:", result.group())


```

##### 2. Extract Data

```python
# When Extracting data we can use re.findall() to find all occurrences of a pattern in a string
import re

text = "Emails: janelleakinyi@example.com, veronicaroth@example.com"
emails = re.findall(r"\S+\S+", text)
print(text)
```

##### 3. Replace Patterns

```python
# When Replacing Text we can use re.sub() to replace patterns with new text
import re

text = "Call me at 0709382707" 
new_text = re.sub(r"\d", "*", text)
print(new_text)

text = "Call me at 0709382707" 
new_text = re.sub(r"07(\d{8})", r"07********", text)
print(new_text)

text = "Call me at 0709382707" 
new_text = re.sub(r"(07\d{2})(\d{4}\d{2})", r"\1****\2", text)
print(new_text)


```

##### 4. Validate Input

```python
# re.match() - Check if the beginning of a string matches a pattern
# It returns a match object if the pattern matches at the start of the string, otherwise it returns None.
import re   

text = "Hello, World!"
match = re.match(r"Hello", text)
if match:
    print("Match found:", match.group())
else:
    print("None")
```

```python
import re

text = "path\to\new\folder"

# title: BAD: Python interprets \n as newline and \t as tab

# This search will fail or behave unpredictably
match1 = ("path\to\new\folder")

# GOOD: Python preserves the string exactly as typed
match2 = (r"path\to\new\folder")

print(match1)

print(match2)
```

```python
import re

text = (r"path\to\new\folder")     

# This search will fail or behave unpredictably due to escape sequences
match1 = re.search("path\to\new\folder", text)
print(match1)

# This search will succeed because the raw string preserves the backslashes
# The pattern is correctly interpreted as path\to\new\folder
match2 = re.search(r"path\\to\\new\\folder", text)
print(match2.group())

```

### `PATTERN`   -  `MEANING`

`\d`    - digits

`\D`    - not a digit

`\w`    - letters and numbers

`\W`    - not a word character

`\s`    - space characters

`\S`    - non-space characters

`.`     - any character

`+`     - 1 or more characters

`*`     - 0 or more characters

`?`     - optional

`^`     - start

`$`     - end

### Character Sets
