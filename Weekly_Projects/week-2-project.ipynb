## `GUESS THE NUMBER`

```python
# This program is a guessing game that generates a number and the user tries to guess it

import random

# Generate a random number between 1 and 100
# store it in x
x = random.randint(1, 100)

# Welcome the user
print("Welcome To Guess The Number!")

# Ask user to choose a level
print("Choose a level")
print(" EASY (10 attempts),\n MEDIUM (7 attempts),\n HARD (5 attempts)\n")

level = input("Enter level: ").upper()   

# Set attempts based on the chosen level
if level == "EASY":
    remaining_attempts = 10
elif level == "MEDIUM":
    remaining_attempts = 7
elif level == "HARD":
    remaining_attempts = 5
else:
    print("Invalid choice! Defaulting to EASY")
    remaining_attempts = 10

used_attempts = 0

print("Guess a number between 1 and 100")

# Start Game loop
while remaining_attempts> 0:
    guess = int(input("Enter a number: "))  

    used_attempts += 1
    remaining_attempts -= 1

    # Check guess
    if guess < x:
        print("Too Low!")
    elif guess > x:
        print("Too High!")
    else:
        print(f"CORRECT! The number is {x}")
        print(f"You guessed it in {used_attempts} attempts.")
        break  # End Game when correct

    print(f"Attempts left: {remaining_attempts}")

# No attempts left means Game is over
if remaining_attempts == 0:
    print(f"GAME OVER! The number is {x}")
    print(f"You guessed it in {used_attempts} attempts")
```
