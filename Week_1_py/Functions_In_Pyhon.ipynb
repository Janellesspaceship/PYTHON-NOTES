## User Defined Functions

- We create User Defined functions using the `def` keyword

Syntax for Functions:

`def` function_name(`parameter(s) names`): # Parameters are also called

arguements `logic`

```python
def square(value):    # value = parameter/ arguement
    squared_value = value ** 2
    print(squared_value)
```

```python
square(23)
```

```python
def modulus(value): 
    '''Returns the remainder of a value when value is divided by 2.'''   
    remainder_value = value % 2
    return remainder_value

num = modulus(23)
print(num)
```

- Passing multiple parameters

```python
def power(base, exponent):
    '''Returns the result of base raised to the power of exponent.'''
    raised_value = base ** exponent
    return raised_value
```

```python
result = power(10, 3)
print(result)
```

## Error Handling


```python
def make_a_list_of_ones(length):
    """ Returns a list of ones of a specified length """
    if length <= 0:
        raise ValueError("Invalid Length. Length must be a positive integer.")
    return [1] * length
```

```python
make_a_list_of_ones(7)
```
