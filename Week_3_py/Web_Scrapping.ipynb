```python
import requests
from bs4 import BeautifulSoup


url = "https://webscraper.io/test-sites/e-commerce/allinone/phones"

# Define headers to mimic a real browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Make the GET request to the URL with the specified headers
response = requests.get(url, headers=headers)


soup = BeautifulSoup(response.content, "html.parser")

products = soup.find_all("a", class_="title")
for product in products:
    print(product.text.strip())
    
```

```python
import requests
from bs4 import BeautifulSoup


url = "https://www.kenyans.co.ke/"


response = requests.get("https://www.kenyans.co.ke/")

soup = BeautifulSoup(response.content, "html.parser")

titles = soup.find_all("h2" ,class_="news-title")

for title in titles[:3]:
    print(title.get_text().strip())


```
