```python
# Inheritance - Reusing and Extending

# Inheritance lets a child class reuse everything from a parent class and add or override as needed.

# This enforces the DRY principle (Don't Repeat Yourself)

# ----- PARENT (BASE) CLASS ----------
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.is_alive = True

    def eat(self):
        print(f"{self.name} is eating.")

    def sleep(self):
        print(f"{self.name} is sleeping.")

    def __str__(self):
        return f"{self.__class__.__name__}: {self.name}, Age {self.age}"
        
# ------CHILD CLASS------------------------
class Dog(Animal):     # Dog inherits everything from Animal
    def __init__(self, name, age, breed):
        super().__init__(name, age)  # Call parent's __init__first
        self.breed = breed           # Add Dog-specific attribute

    def bark(self):                  # New method unique to Dog
        print(f"{self.name} says: Woof!") 

    def fetch(self, item):
        print(f"{self.name} fetch the {item}!")

class Cat(Animal):
    def __init__(self, name, age, is_indoor):
        super().__init__(name, age)
        self.is_indoor = is_indoor

    def purr(self):
        print(f"{self.name} purrs contentedly.")

    def sleep(self):    # OVERIDE parent method wih new behaviour
        super().sleep()            # still call parent's version first
        print(f"   ...{self.name} sleeps for 16 hours straight.") 


# ----MULTI-LEVEL INHERITANCE--------------
class GuideDog(Dog):    # GuideDog inherits from Dog which inherits from Animal
    def __init__(self, name, age, breed, owner):
        super().__init__(name, age, breed)
        self.owner = owner


    def guide(self):
        print(f"{self.name} carefully guides {self.owner} across the street.")
    
```

```python
# Testing
dog2 = Dog("Dixie", 3, "chihuahua")
dog2.bark()
dog2.eat()
print(dog2)
```

```python
dog3 = Dog("Peach", 5, "poodle")
dog3.bark()
dog3.eat()
print(dog3)
```

Abstraction in Python means:

“I will create a rule/template, and force other classes to follow it.”

### How to Achieve Abstraction in Python

- Unlike Java, Python does not have built-in abstract classes by default. We must import the abc module (Abstract Base Classes).

- The Three Rules of Abstraction:

1. Inheritance: You must inherit from ABC.
2. Decorator: You must use `@abstractmethod` on methods you want to enforce.
3. No Instantiation: You cannot create an object from an Abstract Class. If you try, Python will   throw an error.

```python
# 1: The Automobile Standard

# In this example, Car is an abstract concept. You cannot buy a generic "Car"; you must buy a specific brand

# Import the necessary tools
from abc import ABC, abstractmethod

# 2. Define the Abstract Base Class
class Car(ABC):
    @abstractmethod
    def mileage(self): # This acts as a strict rule.
        pass           # Any subclass MUST implement this method, or Python will crash.




# Create concrete classes
class Tesla(Car):
    def mileage(self):
        print("The mileage is 30Kmh (Electric)")

class Suzuki(Car):
    def mileage(self):
        print("The mileage is 25Kmh (Petrol)")

class Duster(Car):
    def mileage(self):
        print("The mileage is 24Kmh (Diesel)")

class Renault(Car):
    def mileage(self):
        print("The mileage is 27Kmh (Hybrid)")

## Execution
# Trying to create the abstract class directly WILL FAILL
# c = Car()
# Error: Can't instatiate abstract class Car with abstract method mileage


# Creating Concrete Objects
t = Tesla()
s = Suzuki()
d = Duster()
r = Renault()

# Calling the methods

t.mileage()
s.mileage()
d.mileage()
r.mileage()
```

## ENCAPSULATION

- Hididng data and controlling access to it inside a class - controlling how its accessed using methods

```python
# Public variables - can be accessed anywhere/modified anywhere
# /no restrictions
# You can use s1.name to change names from bianca to stefan to joy directly

class Student:
    def __init__(self,name):
        self.name = name
s1 = Student("Bianca")
print(s1.name)
s1.name = "Stefan"
print(s1.name)
s1.name = "Joy"
print(s1.name)        
```

Public variables are good for data that does not need protection like username , 
fav color, course name 

disadvantage:

- Anyone can put invalid data though eg

```python
class BankAccount:
    def __init__(self):
        self.balance = 1000
acc = BankAccount()
acc.balance = - 900000    # violates the integrity of a bank account
print(acc.balance)        # Python allows it but its wrong logically

# Why encapsulation exists
```

```python
# without encapsuation
# public variables - default way

class Student:
    def __init__(self):
        self.__grade = 0
    def set_grade(self, grade):
        if 0 <=grade <= 100:
            self.__grade = grade   # __grade is hidden internally inside set_grade()
            # and set_grade() contains my conditional (if) statement
        else:
            print("Invalid Grade")

    # get grade
    def get_grade(self):
        return self.__grade
# Create object = s1
s1 = Student()   # from class Student
s1.set_grade(85)

# modification
s1.__grade = 1000     # not a valid grade but Python still prints it because it is a public variable
print(s1.get_grade())

```

General Structure
1. private variable
2. setter method
3. getter method

class ClassName:
    def __init__(self):
        self.__variable = value  
 `private variable - stores important data privately inside object`
    def set_varaiable(self, new_value):  
`setter method - changes variable safely(validation,restrictions,conditions)`
        self.__variable = new_value
    def get_variable(self):
 `getter method see the value without directly accessing private variable`
        return self.__variable

```python
class Bank_account:
    def __init__(self,balance):
        self.__balance = balance  # private variable encapsulation. cannot be accessed outside the class

    def deposit(self,amount):
        self.__balance += amount  # function to add new amount to hidden balance

    def withdraw(self,amount):
        if amount <= self.__balance:   # checks if money is enough
            self.__balance -= amount   # subtract money if enough balance
        else:
            print("Insufficient funds")
        # get method
    def get_balance(self):
        return self.__balance        

```

```python
acc = Bank_account(1000)
acc.deposit(500)
acc.withdraw(300)
print(acc.get_balance())

acc.__balance = 99999
print(acc.get_balance())  # doesn't change real value
```

- __balance is hidden so no one can directly access it instead we use methods like deposit(), withdraw() and get_balance to safely control and access the balance . simple encapsulation explanation

- `Protected variables` are a warning sign for programmers
- Don't use directly unless necessary
- Internal logic
- May change in future

Good for `inheritance` - child classes can use it safely

```python
class Student:
    def __init__(self):
        self._grade = 50   # you can still access and change it so _ is now security
s1 = Student()
print(s1._grade)
s1._grade = 90
print(s1._grade)        
```
