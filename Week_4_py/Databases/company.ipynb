## SQLite Coding Challenge


A small company wants to track its employees and the departments they belong to.

`1. Create a Database and Tables`

Create an SQLite database named company.db.
Create two tables: employees and departments.
Define a foreign key relationship where each employee belongs to a department.

`2. Insert Sample Data`

Add at least 3 departments.
Add at least 5 employees and assign them to different departments.

`3. Perform CRUD Operations`

Read: Retrieve all employees along with their corresponding department names.
Update: Change a specific employee’s department.
Delete: Remove an employee from the database.
