# NoSQL

`Not Only SQL`

It's a type of database designed for:
- Flexible data.
- Fast scaling.
- Huge amounts of information.
- Modern websites/apps.

Unlike SQL databases, NoSQL databases do `not always use tables and rows.`    

## SQL vs NoSQL

`SQL` `-`  `NoSQL`
- Tables                               - Flexible structures

- Rows and columns                     - Documents or key-values or graphs

- Fixed schema                         - Dynamic schema

- Structured data                      - Unstructured data

- SQLite, MySQL                        - MongoDB, Firebase 

- Best for banking                     - Best for apps or social media

- Scalabilty(vertical-big server)      - (Horizontal- more servers)

### REAL-WORLD EXAMPLES

`SQL`

- Best for: `hospitals`, `banks`, `school systems`, `payroll`.

- This is because `relationships matter`.


`NoSQL`

- Best for: `chat apps`, `social media`, `analytics`, `AI apps`, `recommendation systems`.

- This is because `data changes constantly`.

### MAIN TYPES OF NoSQL

`1. Document Databases`

- Stores data like JSON documents.
- This is powerful because different users can have different fields.
- Also no need for identical columns.
- Example: `MongoDB`

Instead of tables, example:


```python
# JSON
{
    "name": "Tatiana",
    "age": 21,
    "mood_score": 7,
    "interests": ["Lifestyle", "Mental Health"]
}
```

`2. Key-Value Databases`

- It's very fast.
- Used for: `caching`, `sessions`, `settings`.
- Example: `Redis`

example:


```python
user123 - "Tony"
theme - "light"
```

`3. Graph Databases`

- Stores relations.
- Used for: `social networks`, `recommendations`.
- Example: `Neo4j`

```python
Peter - friends with - Sarah 
Sarah - follows - Michael
```

`4. Column Databases`

- Stores data by columns instead of rows.
- Used for: `analytics`, `huge datasets`.
- Example: `Cassandra`

```python
# Read (Query Operations)

# .find()

# Find Sukuna

sukuna = students.find_one({"name": "Sukuna"})
print(sukuna)
```

- Common Query Operators
    - $eq : Equal to
    - $gt : Greater than
    - $lt : Less than
    - $gte : Greater than or Equal to
    - $lte : Less than or Equal to
    - $in : Match any value in a list ({"age": {"$in": [14, 15]}})
    - $set : Only change certain field

```python
# Find Many (Get List)
# Find all students with GPA 
# Syntax: { field: { $operator: value } }
high_achievers = students.find({"gpa": {"$gt": 3.0}})

print("High Achievers:")
for student in high_achievers:
    print(f"- {student['name']} (GPA: {student['gpa']})")
```

```python
# Update

# Simple Update ($set)

# Change Takopi's age to 17
students.update_one(
    {"name": "Takopi San"},        # Filter (who?)
    {"$set": {"age": 17}}          # Action (what?)
)
```

```python
# List Modification

students.update_one(
    {"name": "Takopi San"},
    {"$push": {"courses": "Chemistry"}}  # $push appends values to lists
)
```

```python
# Increment ($inc)

# It's Takopi's birthday! Add 1 to his age.
students.update_one(
    {"name": "Takopi San"},
    {"$inc": {"age": 1}}     # $inc : selects and updates values
)
```

```python
# Delete 

# Delete one
students.delete_one({"name": "Jogo"})
```

```python
# Delete many

# Delete everyone with a low GPA
result = students.delete_many({"gpa": {"$lt": 2.0}})
print(f"Expelled {result.deleted_count} student.")
```
